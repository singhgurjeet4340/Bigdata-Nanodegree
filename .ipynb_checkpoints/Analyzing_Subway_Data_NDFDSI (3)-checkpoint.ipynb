### *Exercise 1.2*

Write down a function that takes the list of all names of the files you downloaded in Exercise 1.1 and compile them into one single file. There must be only one header line in the output file. 

For example, if file_1 has:
line 1...
line 2...

and the other file, file_2, has:
line 3...

We must combine file_1 and file_2 into one master file, as follows:

'C/A, UNIT, SCP, DATEn, TIMEn, DESCn, ENTRIESn, EXITSn'
line 1...
line 2...
line 3...
line 4...
line 5...


````python
files=['turnstile_170603.txt','turnstile_170610.txt','turnstile_170617.txt','turnstile_170624.txt']
outfile='master_file.txt'
def create_master_turnstile_file(filenames,output_file):
    with open(output_file,'w')as master_file:
        master_file.write('C/A,UNIT,SCP,STATION, LINENAME, DIVISION, DATEn,TIMEn,DESCn,ENTRIESn,EXITSn\n')
        for filename in filenames:
            with open(filename,'r') as f1:
                cnt=0
                for line in f1:
                    if cnt>0:
                        master_file.write(line)
                    cnt+=1
create_master_turnstile_file(files,outfile)  
````

````python
# python 3.6

import csv
import pandas as pd
import datetime

def get_hourly_entries(df):
        df['ENTRIESn_hourly'] = (df['ENTRIESn'] - df['ENTRIESn'].shift(1)).fillna(1)
        df['ENTRIESn_hourly']=df['ENTRIESn_hourly'].astype(int)
        return df
df=pd.read_csv('master_file.txt')
df.drop(df.columns[3:6], axis=1, inplace=True)   # deleting the columns as per the question output example
df[' DATEn'] = pd.to_datetime(df[' DATEn'], format='%m/%d/%Y').dt.strftime("%m-%d-%y") #changing the date format as per the question output example
df=get_hourly_entries(df)
df.head()
````

### *Exercise 2.2*

Now, create a function that calculates the number of rainy days. For this, return the count of the number of days where the column *"rain"* is equal to 1.

Tip: You might think that interpreting numbers as integers or floats might not
     work at first. To handle this issue, it might be useful to convert
     these numbers into integers. You can do this by writting cast (column as integer).
     So, for example, if we want to launch the column maxtempi as an integer, we have to
     write something like cast (maxtempi as integer) = 76, instead of just
     where maxtempi = 76.

````python
# python 3.6

import pandas as pd

def num_rainy_days(df):
    count=0
    
    df['rain']=df['rain'].astype(int)
    for index, row in df.iterrows():
        if row['rain']==1:
            count+=1
    return count        
    
filename = "turnstile_data_master_with_weather.csv"
df=pd.read_csv(filename, sep=",")
days=num_rainy_days(df)

print(days)  
````

### *Exercise 2.3*

Calculate if the day was cloudy or not (0 or 1) and the maximum temperature for fog (i.e. the maximum temperature 
     for cloudy days).

````python
# python 3.6

import pandas as pd

def max_temp_aggregate_by_fog(df):
    l=[]
    for index, row in df.iterrows():
        if(row["fog"]==1):
            l.append(row['maxtempi'])
           
    m=max(l)        
    return m        
    
filename = "turnstile_data_master_with_weather.csv"
df=pd.read_csv(filename, sep=",")
maxtemp=max_temp_aggregate_by_fog(df)

print(maxtemp)  # answer=81.0

````

### *Exercise 2.4

Now, calculate the mean for 'meantempi' for the days that are Saturdays or Sundays (weekend):

````python
# python 3.6

import pandas as pd
import datetime
from datetime import datetime

def avg_weekend_temperature(filename):
    mean_temp_weekends =0
    count=0
    
    df=pd.read_csv(filename, sep=",")
    df['DATEn']=pd.to_datetime(df['DATEn'])
    for index, row in df.iterrows():
        if(row['DATEn'].weekday()==5 or row['DATEn'].weekday()==6):
            mean_temp_weekends +=row["meantempi"]
            count+=1
    mean_temp_weekends =mean_temp_weekends/count   
    return mean_temp_weekends
    
filename = "turnstile_data_master_with_weather.csv"
mean=(avg_weekend_temperature(filename))

print(mean)   
````

### *Exercise 2.5

Calculate the mean of the minimum temperature 'mintempi' for the days when the minimum temperature was greater that 55 degrees:

````python
# python 3.6

import pandas as pd
def avg_min_temperature(filename):
    avg_min_temp_rainy=0
    count=0
    df =pd.read_csv(filename, sep=",")
    for index, row in df.iterrows():
        if(row["mintempi"]>55):
            avg_min_temp_rainy+=row["mintempi"]
            count+=1
            
    avg_min_temp_rainy=avg_min_temp_rainy/count        
            
        
    return avg_min_temp_rainy

filename = "turnstile_data_master_with_weather.csv"
mean=avg_min_temperature(filename)

print(mean) 
````

### *Exercise 2.8

Build a function that returns:

1. The mean of entries when it's raining
2. The mean of entries when it's not raining




````python
# python 3.6

import numpy as np

import pandas as pd

def means(turnstile_weather):
    count1=0
    count2=0
    with_rain_mean=0
    without_rain_mean=0
    for index, row in turnstile_weather.iterrows():
        if(row["rain"]==1):
            with_rain_mean+=row['ENTRIESn_hourly']
            count1+=1
        elif row['rain']==0:
            without_rain_mean+=row['ENTRIESn_hourly']
            count2+=1
    with_rain_mean=with_rain_mean/count1
    without_rain_mean=without_rain_mean/count2
    
    return with_rain_mean,without_rain_mean

filename = "turnstile_data_master_with_weather.csv"
turnstile_weather=pd.read_csv(filename, sep=",")
m1,m2=means(turnstile_weather)

print(m1)
print(m2)
    
````

````python
# python 3.6

import sys
sys.stdin = open('turnstile_data_master_with_weather.csv')
sys.stdout = open('mapper_result.txt', 'w')

def mapper():

    for line in sys.stdin:
        data=line.strip().split(",")
        if len(data)==22 and data[1]!='UNIT':
            unit,entries_hourly=data[1],data[6]
            print(unit,"\t",entries_hourly)

mapper()
````

### *Exercise 3.2*

Now, create the reducer. Given the mapper result from the previous exercise, the reducer must print (not return) one line per UNIT, with the total number of ENTRIESn_hourly during May (which is our data duration), separated by a tab. An example of exit line from the reducer may look like this: 'R001 \ t500625.0'

You can assume that the entry for the reducer is ordered in a way that all lines corresponding to a particular unit are grouped. However, the reducer exit will have repetition, as there are stores that appear in different files' locations.

Export your reducer into a file named reducer_result.txt and send it with your submission.

````python
# python 3.6

import sys
sys.stdin = open('mapper_result.txt')
sys.stdout = open('reducer_result.txt', 'w')

def reducer():
    
    hourTotal=0
    oldKey=None
    for line in sys.stdin:
        mapped_data=line.strip().split('\t')
        if len(mapped_data)!=2:
            continue
        Key,Hour=mapped_data
        Hour=float(Hour)
        if oldKey and oldKey!=Key:
            print(oldKey,'\t',hourTotal)
            oldKey=Key
            hourTotal=0
        oldKey=Key
        hourTotal+=float(Hour)
        
    if oldKey!=None:
        print(oldKey,'\t',hourTotal)

        
reducer()
````
